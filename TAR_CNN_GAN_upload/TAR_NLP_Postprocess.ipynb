```python
# Postprocess
#Post process training results 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

```

```python
def epoch_plot(logFile):
    frame = pd.read_csv(logFile)
    frame.plot(x = 'epoch', y = ' val_Acc', figsize = (10,6.18))
    print("max val acc occurs at epoch_"+str(frame[' val_Acc'].idxmax()))
    print("max val acc: "+str(max(frame[' val_Acc'])))

def G_loss_plot(logFile):
    frame = pd.read_csv(logFile)
    #frame.plot(x = 'epoch', y = ' train_loss_D', figsize = (10,6.18))
    frame.plot(x = 'epoch', y = ' train_loss_G', figsize = (10,6.18))

def D_loss_plot(logFile):
    frame = pd.read_csv(logFile)
    frame.plot(x = 'epoch', y = ' train_loss_D', figsize = (10,6.18))
```

```python
filename = "SSL_GAN_log_NLP_embedding_80.csv"
```

```python
epoch_plot(filename)
```

```python
D_loss_plot(filename)
```

```python
G_loss_plot(filename)
```
